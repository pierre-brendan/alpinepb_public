```python
# import packages
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import time
import datetime
```

```python
# add the crawl date
current_date = datetime.datetime.now()
filename = str(current_date.day)+str(current_date.month)+str(current_date.year)
```

```python
# Weather
import urllib.request, json 
with urllib.request.urlopen("https://ads.rubiconproject.com/floors/10738-pbjs-floors.json") as url:
    data = json.loads(url.read().decode())
    print(data)
dfb = pd.json_normalize(data, 'modelGroups')
dfb['publisherDomain'] = 'www.weather.com'
dfb['crawl_date'] = current_date
dfb.to_csv(str(filename + '_weather_floor_data.csv'), index=False)
#dfb.to_csv('weathers_floor_model.csv', index=False)
```
